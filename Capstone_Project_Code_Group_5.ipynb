```python

data['lpep_pickup_datetime'] = pd.to_datetime(data['lpep_pickup_datetime'])
data['lpep_dropoff_datetime'] = pd.to_datetime(data['lpep_dropoff_datetime'])
# Verify the data types after conversion
print(data.dtypes)
```

```python
# Counting the unique locations
num_locations = data['PULocationID'].nunique()
print("Number of unique locations:", num_locations)
```

```python
# Checking for the duplicate rows in the DataFrame
duplicate_rows = data[data.duplicated()]

if duplicate_rows.empty:
    print("No duplicate rows found.")
else:
    print("Duplicate rows found:")
    print(duplicate_rows)
```

```python
# Checking for the negative values in numerical columns of the DataFrame
numerical_columns = data.select_dtypes(include=['number']).columns
columns_with_negatives = data[numerical_columns].columns[(data[numerical_columns] < 0).any()]

if not columns_with_negatives.empty:
    print("The following columns contain negative values:")
    print(columns_with_negatives)
else:
    print("No negative values found in any numerical column.")
negative_value_counts = {}

# Counting the number of negative values for each column
for col in numerical_columns:
    num_negative_values = (data[col] < 0).sum()
    negative_value_counts[col] = num_negative_values
for col, count in negative_value_counts.items():
    print(f"{col}: {count} negative values")
```

```python
# columns to absolutize
columns_to_absolutize = ['fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge', 'total_amount', 'congestion_surcharge']
data[columns_to_absolutize] = data[columns_to_absolutize].abs()
```

```python
# Calculating percentage of missing values per column
missing_values_percentage = (data.isnull().sum() / len(data)) * 100
print("Percentage of missing values per column:")
print(missing_values_percentage)
missing_rows_percentage = (data.isnull().any(axis=1).sum() / len(data)) * 100
print("\nPercentage of missing rows in the DataFrame:", missing_rows_percentage)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import seaborn as sns
# Calculating the total fare including all charges
data['total_fare'] = data['fare_amount'] + data['extra'] + data['mta_tax'] + data['tip_amount'] + data['tolls_amount'] + data['improvement_surcharge'] + data['congestion_surcharge']

# Calculate total fare including all charges
#data['total_fare'] = data['fare_amount'] + data['extra'] + data['mta_tax'] + data['tip_amount'] + data['tolls_amount'] + data['improvement_surcharge'] + data['congestion_surcharge']

# Spatial Analysis
pickup_neighborhoods = data.groupby('PULocationID')[['trip_duration', 'total_fare']].mean().reset_index()
dropoff_neighborhoods = data.groupby('DOLocationID')[['trip_duration', 'total_fare']].mean().reset_index()

# Temporal Analysis
data['pickup_hour'] = data['lpep_pickup_datetime'].dt.hour
data['pickup_day_of_week'] = data['lpep_pickup_datetime'].dt.day_name()

hourly_congestion = data.groupby('pickup_hour')[['trip_duration', 'total_fare', 'congestion_surcharge']].mean().reset_index()
daily_congestion = data.groupby('pickup_day_of_week')[['trip_duration', 'total_fare', 'congestion_surcharge']].mean().reset_index()

# Correlation with congestion surcharge
#congestion_surcharge_correlation = data[['trip_duration', 'trip_distance', 'total_fare', 'congestion_surcharge']].corr()


print("Top pickup neighborhoods by average trip duration:")
print(pickup_neighborhoods.sort_values('trip_duration', ascending=False).head(10))
print("\nTop pickup neighborhoods by average fare:")
print(pickup_neighborhoods.sort_values('total_fare', ascending=False).head(10))

print("\nTop dropoff neighborhoods by average trip duration:")
print(dropoff_neighborhoods.sort_values('trip_duration', ascending=False).head(10))
print("\nTop dropoff neighborhoods by average fare:")
print(dropoff_neighborhoods.sort_values('total_fare', ascending=False).head(10))

print("\nHourly variations in congestion levels (average trip duration, fare, and congestion surcharge):")
print(hourly_congestion)

print("\nDaily variations in congestion levels (average trip duration, fare, and congestion surcharge):")
print(daily_congestion)

#print("\nCorrelation matrix:")
#print(congestion_surcharge_correlation)
# EDA
# Visualize hourly variations
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.lineplot(x='pickup_hour', y='trip_duration', data=hourly_congestion, label='Trip Duration')
plt.xlabel('Hour of Day')
plt.ylabel('Average Trip Duration (mins)')
plt.title('Hourly Trip Duration')

plt.subplot(1, 3, 2)
sns.lineplot(x='pickup_hour', y='total_fare', data=hourly_congestion, label='Total Fare')
plt.xlabel('Hour of Day')
plt.ylabel('Average Total Fare ($)')
plt.title('Hourly Total Fare')

plt.subplot(1, 3, 3)
sns.lineplot(x='pickup_hour', y='congestion_surcharge', data=hourly_congestion, label='Congestion Surcharge')
plt.xlabel('Hour of Day')
plt.ylabel('Average Congestion Surcharge ($)')
plt.title('Hourly Congestion Surcharge')

plt.tight_layout()
plt.show()
```

# Using Ridge Regression (RFE) for Feature Selection

```python
features = data[['trip_distance','pickup_hour', 'congestion_surcharge', 'congestion_neighborhood','pickup_day_of_week','PULocationID','trip_duration']]
target = data['total_fare']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
from sklearn.feature_selection import RFE
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1.0)
rfe = RFE(ridge, n_features_to_select=4)
rfe.fit(X_train, y_train)
print("\nSelected Features:")
for feature, importance in zip(features.columns, rfe.ranking_):
    if importance == 1:
        print(f"{feature}")
X_train_selected = rfe.transform(X_train)
X_test_selected = rfe.transform(X_test)
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Linear Regression Score
linear_regression = LinearRegression()
linear_regression.fit(X_train_selected, y_train)
linear_regression_score = linear_regression.score(X_test_selected, y_test)
print(f"\nLinear Regression Score: {linear_regression_score:.2f}")

# Decision Tree Regression Score
decision_tree = DecisionTreeRegressor(random_state=42)
decision_tree.fit(X_train_selected, y_train)
decision_tree_score = decision_tree.score(X_test_selected, y_test)
print(f"Decision Tree Regression Score: {decision_tree_score:.2f}")

# Random Forest Regression Score
random_forest = RandomForestRegressor(random_state=42)
random_forest.fit(X_train_selected, y_train)
random_forest_score = random_forest.score(X_test_selected, y_test)
print(f"Random Forest Regression Score: {random_forest_score:.2f}")
```

#Evaluation Metrices:

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Linear Regression Evaluate metrics:
y_pred_linear = linear_regression.predict(X_test_selected)
linear_mse = mean_squared_error(y_test, y_pred_linear)
linear_r2 = r2_score(y_test, y_pred_linear)
linear_mae = mean_absolute_error(y_test, y_pred_linear)
print(f"\nLinear Regression Metrics:")
print(f"Mean Squared Error: {linear_mse:.2f}")
print(f"R-squared: {linear_r2:.2f}")
print(f"Mean Absolute Error: {linear_mae:.2f}")

# Decision Tree Regression Evaluate metrics:
y_pred_decision_tree = decision_tree.predict(X_test_selected)
decision_tree_mse = mean_squared_error(y_test, y_pred_decision_tree)
decision_tree_r2 = r2_score(y_test, y_pred_decision_tree)
decision_tree_mae = mean_absolute_error(y_test, y_pred_decision_tree)
print(f"\nDecision Tree Regression Metrics:")
print(f"Mean Squared Error: {decision_tree_mse:.2f}")
print(f"R-squared: {decision_tree_r2:.2f}")
print(f"Mean Absolute Error: {decision_tree_mae:.2f}")

# Random Forest Regression Evaluate metrics
y_pred_random_forest = random_forest.predict(X_test_selected)
random_forest_mse = mean_squared_error(y_test, y_pred_random_forest)
random_forest_r2 = r2_score(y_test, y_pred_random_forest)
random_forest_mae = mean_absolute_error(y_test, y_pred_random_forest)
print(f"\nRandom Forest Regression Metrics:")
print(f"Mean Squared Error: {random_forest_mse:.2f}")
print(f"R-squared: {random_forest_r2:.2f}")
print(f"Mean Absolute Error: {random_forest_mae:.2f}")

# Linear Regression Vs Actual Fares:
y_pred_linear = linear_regression.predict(X_test_selected)
print("\nLinear Regression Predicted vs Actual Fares:")
for i in range(10):
    print(f"Predicted: {y_pred_linear[i]:.2f}, Actual: {y_test.iloc[i]:.2f}")

# Decision Tree Regression Vs Actual Fares:
y_pred_decision_tree = decision_tree.predict(X_test_selected)
print("\nDecision Tree Regression Predicted vs Actual Fares:")
for i in range(10):
    print(f"Predicted: {y_pred_decision_tree[i]:.2f}, Actual: {y_test.iloc[i]:.2f}")

# Random Forest Regression Vs Actual Fares:
y_pred_random_forest = random_forest.predict(X_test_selected)
print("\nRandom Forest Regression Predicted vs Actual Fares:")
for i in range(10):
    print(f"Predicted: {y_pred_random_forest[i]:.2f}, Actual: {y_test.iloc[i]:.2f}")
```
